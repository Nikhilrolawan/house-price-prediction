```python
import matplotlib.pyplot as plt
import numpy as np
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
x_train.shape[0]
plt.scatter(x_train,y_train,marker='x',c='r')
plt.ylabel('price(in 1000s)')
plt.xlabel('house size in 1000 sq ft')
plt.show()
```

```python
def compute_model(x,w,b):# x : the house size array , w and b are parametres
    m=x.shape[0] # shape of the array in this case 2
    f_wb=np.zeros(m) # will create the 2 shape array with all entries 0
    for i in range(m):
        f_wb[i]=w*x[i]+b # to store the data in 2 shape array
    return f_wb

data=compute_model(x_train,200,100) #chossing the right value of w and b . try different values of w and b to match 
#our prediction with actual values
plt.plot(x_train,data,label='our prediction',c='r')
plt.scatter(x_train,y_train,label='actual values',c='b')
plt.title('Housing Price')
plt.ylabel('price(in 1000s)')
plt.xlabel('house size in 1000 sq ft')
plt.legend()
plt.show()
```
